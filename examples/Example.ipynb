```python
import numpy as np
from LadrunoGraphStyle import set_default_plot_params, main_colors
set_default_plot_params()



from WaveLadruno.tools.moment_tensor_tools import moment_tensor_from_strike_dip_rake
from WaveLadruno.plots import plot_radiation_patterns, plot_3d_colored_sphere, plot_sphere_with_vectors, plot_beachball_set
from WaveLadruno.tools.radiation_vectors import get_all_patterns
```

```python
# === √Ångulos ===
theta = np.deg2rad(45)
phi   = np.deg2rad(30)

# === Vector gamma ===
gamma = np.array([
                    np.sin(theta)*np.cos(phi),
                    np.sin(theta)*np.sin(phi),
                    np.cos(theta)
])

get_all_patterns(gamma)
```

```python
# M = moment_tensor_from_strike_dip_rake(strike=0, dip=90, rake=0, M0=1)
M = moment_tensor_from_strike_dip_rake(strike=62, dip=81, rake=-151, M0=1)
print(np.round(M,3))
```

```python
# === Ejemplo: far-field P y S ===
plot_radiation_patterns(M , "AFP", projection='r')       # (a) P-wave
# plot_radiation_patterns(M , "AFS", projection='S')       # (b) S total
plot_radiation_patterns(M , "AFS", projection='theta')   # (c) SV-wave
plot_radiation_patterns(M , "AFS", projection='phi')     # (d) SH-wave
```

```python
plot_beachball_set(M, mechanism_name="Beach Ball 2D")
```

```python
# Beach ball 3D
plot_3d_colored_sphere("AFP", M, projection='r', title="P-wave Beach Ball 3D")
# plot_3d_colored_sphere("AFS", M, projection='S', title="S-wave Beach Ball 3D")
plot_3d_colored_sphere("AFS", M, projection='theta', title="SV-wave Beach Ball 3D")
plot_3d_colored_sphere("AFS", M, projection='phi', title="SH-wave Beach Ball 3D")
```

```python
# Beach ball 3D with vectors
plot_sphere_with_vectors("AFP", M, projection='r', title="P-wave:  Beach Ball 3D + Vectors")
# plot_sphere_with_vectors("AFS", M, projection='S', title="S-wave  Beach Ball 3D + Vectors")
plot_sphere_with_vectors("AFS", M, projection='theta', title="SV-wave:  Beach Ball 3D + Vectors")
plot_sphere_with_vectors("AFS", M, projection='phi', title="SH-wave:  Beach Ball 3D + Vectors")
```
